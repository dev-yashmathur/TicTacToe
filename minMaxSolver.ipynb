```python
rows = data.shape[0]
for i in range(7,0,-1): #column name
    prev = str(i+1) + "_res"
    col_res = []
    int_j = 0 #row number
    while int_j < rows:
        j = data.at[int_j, i]
        if j == 0: #Carry forward result
            col_res.append(data.at[int_j, prev])
            int_j += 1
            continue
        # int_j+=1
        # sizeOfSim = 1 #Since atleast 1 will be similar
        
        if data.at[int_j, i+1] == 0: #this is the decisive move
            col_res.append(data.at[int_j, prev])
            int_j += 1
            continue
        
        tempRes = []
        tempRes.append(data.at[int_j,prev])
        start = int_j
        int_j += 1
        while int_j<rows and data.at[int_j, i] == data.at[int_j-1, i]:
            # sizeOfSim +=1
            tempRes.append(data.at[int_j,prev]) #Getting the results of all similar series
            int_j+=1
        
        score = 0
        if i%2 == 0: #Player 2's turn
            if 1 in tempRes: #because next turn is player 1's
                score = 1
            elif 0.5 in tempRes:
                score = 0.5
            else:
                score = -1
        else: #Player 1's
            if -1 in tempRes:
                score = -1
            elif 0.5 in tempRes:
                score = 0.5
            else:
                score = 1

        
        
        moveScore = [score for _ in tempRes]
        col_res.extend(moveScore)

    current = str(i) + "_res"
    data[current] = col_res


data
        
            # Get a list of all the results for series of moves uptil 7
            # check if there exists +1 if current move is player 1 and vice versa
            # for the size of all series (by comment 1) fill it with +1, -1 or 0
            # append this list to the col_res
            
```
