```python
import itertools
perms = itertools.permutations([1,2,3,4,5,6,7,8,9])
data = pd.DataFrame(columns = [1,2,3,4,5,6,7,8,9])
data_list = []
for i in list(perms):
    dic_rep = {j : i[j-1] for j in range(1,10)}
    # print(dic_rep)
    data_list.append(dic_rep)

data = pd.DataFrame.from_dict(data_list)

data.describe()
```

```python
import tictactoe
mini = pd.DataFrame(data.iloc[[x for x in range(15)]])
results = []
for i in range(len(mini)):
    flag = False
    Test = tictactoe.Board()
    x = 0
    int_j = 0
    for j in mini.iloc[i]:
        int_j+=1 #the index from which the others have to be 0
        Test.play(j, (x%2)+1)
        x+=1
        res = Test.check_win()
        if res[0] == None:
            Test.display()
            print("\n\n")
            flag = True
            results.append(0.5)
            break
        elif res[0] == True:
            Test.display()
            print("\n\n")
            flag = True
            results.append(1)
            break
        
    if flag == True:
        for y in range(int_j, 9):
            mini.iat[i, y] = 0

print(results)
mini
```

```python
results = []
for i in range(len(data)):
    Test = tictactoe.Board()
    flag = False
    x = 0
    int_j = 0
    for j in data.iloc[i]:
        int_j+=1
        Test.play(j, (x%2)+1)
        x+=1
        res = Test.check_win()
        if res[0] == None:
            # Test.display()
            # print("\n\n")
            results.append(0.5)
            flag = True
            break
        elif res[0] == True:
            # Test.display()
            # print("\n\n")
            if res[1] == 1:
                results.append(1)
            else:
                results.append(-1)
            flag = True
            break
    if flag == True:
        for y in range(int_j, 9):
            data.iat[i, y] = 0

print(len(results))
```
